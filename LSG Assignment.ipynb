```python
edge_betweenness=nx.edge_betweenness_centrality(G)
edge_betweenness
```

```python
def max_centrality_edge(G):
    ed = nx.edge_betweenness_centrality(G)
    sorted_edges_cent = sorted(ed.items(), key=lambda x:x[1], reverse=True)
    edg = sorted_edges_cent[0]
    return edg[0]
```

```python
def partition(G):
    removed_edges=[]
    print("Edges to remove:", end="")
    while len(list(nx.connected_components(G)))==1:
        edge=max_centrality_edge(G)
        G.remove_edge(*edge)
        removed_edges.append(edge)
        print(edge, end="")
        return [G, removed_edges]

rao=0
AP=0
while rao!=2:
    AP=AP+1
    k=partition(G)
    print("no.of edges removed:",len(k[1]))
    G1=k[0]
    rao=len(list(nx.connected_components(G1)))
    print("no. of connected components:",len(list(nx.connected_components(G1))))
    print("total edges removed = ",AP)
pos=nx.spring_layout(G1)
nx.draw(G1,pos=pos,node_size=1)
```
